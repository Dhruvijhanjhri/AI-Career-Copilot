```python
import pandas as pd
import numpy as np
import re
```

```python
resume_df = pd.read_csv("../data/AI_Resume_Screening.csv")

resume_df.head()
```

```python
resume_df.info()
```

```python
resume_df.isnull().sum()
```

```python
resume_df["Certifications"] = resume_df["Certifications"].fillna("Not Provided")
```

```python
resume_df.isnull().sum()
```

```python
resume_df = resume_df.drop_duplicates()
```

```python
resume_df.shape
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z0-9 ]', '', text)
    return text
```

```python
resume_df.to_csv(
    "../data/cleaned_resumes.csv",
    index=False
)
```

```python
jobs_df = pd.read_csv("../data/fake_job_postings.csv")
```

```python
jobs_df.head()
```

```python
jobs_df = jobs_df.dropna()
jobs_df = jobs_df.drop_duplicates()
```

```python
jobs_df["description"] = jobs_df["description"].astype(str).apply(clean_text)
```

```python
jobs_df.to_csv(
    "../data/cleaned_jobs.csv",
    index=False
)
```
